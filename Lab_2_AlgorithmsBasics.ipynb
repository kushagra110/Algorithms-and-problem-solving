```python
# Exercise 2
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result
```

```python
# Sample function call: func2(2, 4)
import time
start=time.time()
print (func2(2,4))
end = time.time()
ellapsed_time = (end-start)*1000
print (ellapsed_time)
# the function is performing a to the power b
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Sample function call: func4([1, 2, 3, 4], [2, 4, 6, 8])
import time
start=time.time()
print (func4([1, 2, 3, 4], [2, 4, 6, 8]))
end = time.time()
ellapsed_time = (end-start)*1000
print (ellapsed_time)
# the function is first sorting the second array and then doing the intersection to count the same number of elements
```

```python
# Exercise 5
def func5(n):
    guess = 1
    while guess * guess <= n:
        if guess * guess == n:
            return guess
        guess = guess + 1
    return None
```

```python
# Sample function call: func5(16)
import time
start=time.time()
print (func5(16))
end = time.time()
ellapsed_time = (end-start)*1000
print (ellapsed_time)
# the function is giving the no. which is the square of the no.
```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A
```

```python
#Sample function call: func8([3,1,2,7,8,6])
import time
start=time.time()
print (func8([3,1,2,7,8,6]))
end = time.time()
ellapsed_time = (end-start)*1000
print (ellapsed_time)
#swapping the numbers in the array
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

```

```python
#Sample function call: func12([1, 2, 3, 4, 5], 2)
import time
start=time.time()
print (func12([1, 2, 3, 4, 5], 2))
end = time.time()
ellapsed_time = (end-start)*1000
print (ellapsed_time)
# the function is shifting the array at the given position
```

Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
# write a function named as sum_arr to find sum of elements in given array
def sum_arr(a):
    n=len(a)
    sum=0
    for i in range(n):
      sum+=a[i]
    return sum

```

```python
#call the function
arr = [12, 3, 4, 15,45,78,450]
#print(sum_arr)
import time
start=time.time()
print(sum_arr(arr))
end = time.time()
ellapsed_time = (end-start)*1000
print (ellapsed_time)
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  max=arr[0]
  min=arr[0]
  for i in range(len(arr)):
     if arr[i]>max:
        max=arr[i]
     elif arr[i]<min:
        min=arr[i]
  return max,min
```

```python
# call the function
arr = [12, 1234, 45, 67, 1]
import time
start=time.time()
print(max_min(arr))
end = time.time()
ellapsed_time = (end-start)*1000
print (ellapsed_time)
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(a):
  n=len(a)
  e,o=0,0
  for i in range(n):
    if(a[i]%2==0):e=e+1
    else:o=o+1
  print("even=",e)
  print("odd=",o)

```
