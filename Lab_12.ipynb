# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_multiplication(p):
    n = len(p)

    # Create DP table
    m = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (m[i][k] +
                        m[k+1][j] +
                        p[i-1] * p[k] * p[j])

                if cost < m[i][j]:
                    m[i][j] = cost

    return m

```

```python
# Input dimensions
p = [10, 30, 5, 60]

table = matrix_chain_multiplication(p)

print("Minimum multiplication cost:", table[1][len(p)-1])

print("\nDP Table:")
for row in table[1:]:
    print(row[1:])
```

```python
import sys

def matrix_chain_order(p):
    n = len(p) - 1  # number of matrices

    # m[i][j] = minimum cost of multiplying Ai...Aj
    m = [[0] * n for _ in range(n)]

    # s[i][j] = index at which to split for optimal cost
    s = [[0] * n for _ in range(n)]

    for L in range(2, n + 1):  # chain length
        for i in range(n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize

            for k in range(i, j):
                cost = (
                    m[i][k]
                    + m[k + 1][j]
                    + p[i] * p[k + 1] * p[j + 1]
                )

                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k

    return m, s


def print_optimal_parens(s, i, j):
    if i == j:
        return f"A{i+1}"
    else:
        k = s[i][j]
        left = print_optimal_parens(s, i, k)
        right = print_optimal_parens(s, k + 1, j)
        return f"({left}{right})"


def naive_cost(p):
    """
    Left-to-right multiplication cost
    ((((A1A2)A3)A4)...)
    """
    cost = 0
    current_rows = p[0]
    current_cols = p[1]

    for i in range(2, len(p)):
        # multiply current result with next matrix
        cost += current_rows * current_cols * p[i]
        current_cols = p[i]

    return cost


def main():
    # Given input
    p = [5, 10, 3, 12, 5, 50, 6]

    # 1 & 2: Compute minimum cost
    m, s = matrix_chain_order(p)
    min_cost = m[0][len(p) - 2]

    # 3: Optimal parenthesization
    optimal_parens = print_optimal_parens(s, 0, len(p) - 2)

    # 4: Compare with naive
    naive = naive_cost(p)
    saved = naive - min_cost

    print("Matrix dimensions (p):", p)
    print("\nMinimum number of multiplications:", min_cost)
    print("Optimal parenthesization:", optimal_parens)
    print("\nNaive (left-to-right) cost:", naive)
    print("Multiplications saved:", saved)


if __name__ == "__main__":
    main()
```
