# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    predecessors = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')
        predecessors[node] = None

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        # Find the unvisited node with the smallest distance
        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None: # All remaining nodes are unreachable
            break

        visited.append(min_node)

        # Update distances of neighbor nodes
        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node

    return distances, predecessors
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
import collections

# 1. Modify program for user input graph using dictionary.
#    For demonstration, we'll define a new graph directly here,
#    simulating how a user might define it. A truly interactive
#    input would require more complex parsing or a GUI.
# 3. Add two more nodes in graph (E, F).
graph_new = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 6},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
    'D': {'B': 10, 'C': 3, 'E': 7},
    'E': {'B': 6, 'D': 7, 'F': 1},
    'F': {'C': 8, 'E': 1}
}

print("New Graph:")
for node, neighbors in graph_new.items():
    print(f"  {node}: {neighbors}")
print("\n")

# Helper function to reconstruct the shortest path
def get_shortest_path(predecessors, start, target):
    path = collections.deque()
    current = target
    while current is not None:
        path.appendleft(current)
        current = predecessors.get(current)
    if path and path[0] == start: # Check if a valid path from start to target was found
        return list(path)
    else:
        return [] # No path found


# 4. Test algorithm with different source nodes.
source_nodes_to_test = ['A', 'B', 'E']

for source in source_nodes_to_test:
    print(f"--- Testing Dijkstra from Source: {source} ---")
    distances, predecessors = dijkstra(graph_new, source)

    # 2. Display shortest path with total distance.
    for target_node in graph_new:
        if target_node == source:
            print(f"Shortest path from {source} to {target_node}: {source} (Total Distance: 0)")
            continue

        path = get_shortest_path(predecessors, source, target_node)
        total_distance = distances.get(target_node, float('inf'))

        if path and total_distance != float('inf'):
            print(f"Shortest path from {source} to {target_node}: {' -> '.join(path)} (Total Distance: {total_distance})")
        else:
            print(f"No path found from {source} to {target_node} (Total Distance: unreachable)")
    print("\n")
```
